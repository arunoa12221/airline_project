Exploratory Data Analysis (EDA)

```python
# delay % calculation
df['delay_rate']=(df['arr_del15']/df['arr_flights'])*100
df["cancellation_rate"] = (df["arr_cancelled"] / df["arr_flights"]) * 100

# Total flights vs delayed flights
totals=df[['arr_flights','arr_del15']].sum()

plt.bar(["Total Flights", "Delayed Flights"], totals, color=["skyblue", "salmon"])
plt.title('OVERALL FLIGHTS VS DELAYED FLIGHTS',size=13,style='oblique')
plt.ylabel('NUMBER OF FLIGTS',size=13,style='oblique')
plt.show()
```

```python
# Delay % across carriers
plt.figure(figsize=(12,6))
df.groupby("carrier_name")["delay_rate"].mean().sort_values().plot(kind="barh", color="orange",grid=True)
plt.title("Average Delay % Across Carriers", fontsize=14, fontweight="bold")
plt.xlabel("Delay %")
plt.show()

# Delay % across airports
plt.figure(figsize=(12,6))
df.groupby("airport_name")["delay_rate"].mean().sort_values(ascending=False).head(10).plot(kind="bar", color="purple",grid=True)
plt.title("Top 10 Airports by Delay %", fontsize=14, fontweight="bold")
plt.ylabel("Delay %")
plt.xticks(rotation=90, ha="right")
plt.show()
```

(b) Top Problematic Carriers / Airports 
● Group by carrier_name → average delay_rate. 
● Group by airport_name → average delay_rate.

```python
df.columns = df.columns.str.strip()

df["delay_rate"] = (df["arr_del15"] / df["arr_flights"]) * 100
carrier_delay = (df.groupby("carrier_name")["delay_rate"].mean().sort_values(ascending=False).head(10))

print("Top 5 Carriers by Avg Delay %:\n", carrier_delay)

plt.figure(figsize=(12,5))
sns.barplot(x=carrier_delay.values, y=carrier_delay.index, palette="Reds_r")
plt.title("Top 10 Problematic Carriers (Avg Delay %)")
plt.xlabel("Average Delay %")
plt.ylabel("Carrier")
plt.show()
```

```python
airport_delay = (df.groupby("airport_name")["delay_rate"].mean().sort_values(ascending=False).head(10))

print("\nTop 5 Airports by Avg Delay %:\n", airport_delay)

plt.figure(figsize=(12,5))
sns.barplot(x=airport_delay.values, y=airport_delay.index, palette="Blues_r")
plt.title("Top 10 Problematic Airports (Avg Delay %)")
plt.xlabel("Average Delay %")
plt.ylabel("Airport")
plt.show()
```

```python
# select the delay cause columns
cause_cols=['carrier_ct','weather_ct','late_aircraft_ct']

# groupby by carrier (we can also try 'airport_name' instead)
carrier_causes=df.groupby('carrier_name')[cause_cols].sum()

# Find the maximum cause per carrier 
carrier_causes['top_cause']=carrier_causes.idxmax(axis=1)

('Top delay cause per carrier:')
(carrier_causes['top_cause'])

```

```python
colors=sns.color_palette('GnBu',n_colors=len(cause_cols))

ax=carrier_causes[cause_cols].plot(kind='bar',stacked=True,figsize=(12,6),color=colors,edgecolor='black')

plt.title('Delay Causes Breakdown By carrier',fontsize=16,fontweight='bold')
plt.xlabel('Carrier',fontsize=12)
plt.ylabel('Number of Delays',fontsize=12)
sns.despine()
plt.tight_layout()
plt.show()
```

```python
# Calulate cancellation_rate
df['cancellation_rate']=(df['arr_cancelled']/df['arr_flights'])*100

# plot
plt.figure(figsize=(12,6))
sns.barplot(data=df,x='carrier_name',y='cancellation_rate',palette='seismic',estimator='mean')
plt.title('Cancellation Rate by Carrier',size=18,fontweight='bold')
plt.xlabel('Carrier',fontsize=12)
plt.ylabel('cacellation rate(%)',fontsize=12)
plt.xticks(rotation=90,size=12)
plt.tight_layout()
plt.show()
```

2--> Top 5 airports with highest delay_rate (LINE PLOT)

3---> Delay Causes Contribution per Carrier  (STACKED BAR PLOT)

```python
cause_cols=['carrier_ct','weather_ct','nas_ct','late_aircraft_ct']

carrier_causes= df.groupby('carrier_name')[cause_cols].sum()

carrier_causes.plot(kind='bar',stacked=True,figsize=(12,6),colormap='Set2',edgecolor='black')
plt.title('Delay Causes Contribution per Carrier',size=16,fontweight='bold')
plt.xlabel('Carrier',style='oblique',size=14)
plt.ylabel('Number of Delays',style='oblique',size=14)
```

4----> Overall distribution of delay causes  (PIE CHART)

```python
total_causes=df[cause_cols].sum()

plt.figure(figsize=(8,8))
plt.pie(total_causes,labels=total_causes.index,autopct='%1.1f%%',colors=sns.color_palette('Set3'))
plt.title('Overall distribution of delay causes',size=16,fontweight='bold')
```

 Delays (arr_delay) by carrier vs airpor    (SCATTER PLOT)

```python
grp=df.groupby(['carrier_name','airport_name'])['arr_delay'].mean().reset_index()

plt.figure(figsize=(12,6))
sns.scatterplot(data=grp,x='carrier_name',y='arr_delay',hue='airport_name',palette='flare',edgecolor='black',legend=False)
plt.title('Arrival delays by Carrier vs Airport',size=16,fontweight='bold')
plt.xlabel('Carrier',size=14)
plt.ylabel('Arrival Delay',size=14)
plt.xticks(rotation=90)
```

6---->  Monthly delay trend . (HIST PLOT)

```python
monthly_delay=df.groupby('month')['arr_del15'].mean().reset_index()

plt.figure(figsize=(10,5))
sns.histplot(data=monthly_delay, x='month', weights='arr_del15',bins=12,color='orange')
plt.title('Monthly Delay Trend',size=18,fontweight='bold')
plt.xlabel('Month',size=14)
plt.ylabel('Average Delays (>15 min)')
plt.xticks(range(1,13))
plt.show()
plt.savefig('monthly_delay.jpg')
```

Recommendations

The biggest problem is “late aircraft delays” (when the incoming plane arrives late, the next flight also gets delayed).

Airlines should improve turnaround times (prepare planes faster for the next flight).

Airports with frequent delays should increase efficiency in traffic handling.

In simple words: Airlines should manage their planes better, and busy airports should improve operations to avoid delays.
